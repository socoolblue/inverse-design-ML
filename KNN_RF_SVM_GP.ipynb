```python
import tensorflow as tf
import numpy as np
import random
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

tf.reset_default_graph()  
tf.set_random_seed(777)  # for reproducibility
np.random.seed(999)

xy = np.loadtxt("Input_data_path", delimiter=',', dtype=np.float32) #x8, x16

# xy_data=xy.reshape(-1,10) #x8
xy_data=xy.reshape(-1,18) #x16
np.random.shuffle(xy_data)

x_data_training = xy_data[:4500, :-2]
y_data_training = xy_data[:4500, -2]
x_data_test = xy_data[4500:, :-2]
y_data_test = xy_data[4500:, -2:]

knn_reg = KNeighborsRegressor(n_neighbors=5, n_jobs=-1)
knn_reg.fit(x_data_training, y_data_training)
y_pred_tr=knn_reg.predict(x_data_training)
y_pred=knn_reg.predict(x_data_test)

print('{:.4f}'.format(r2_score(y_data_training, y_pred_tr, multioutput='variance_weighted'))) 
print('{:.4f}'.format(r2_score(y_data_test, y_pred, multioutput='variance_weighted')))
print('{:.4f}'.format(mean_squared_error(y_data_training, y_pred_tr, multioutput='uniform_average')))
print('{:.4f}'.format(mean_squared_error(y_data_test, y_pred, multioutput='uniform_average'))) 

#k_nearest neighbors_regression
```

```python
import tensorflow as tf
import numpy as np
import random
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

tf.reset_default_graph()  
tf.set_random_seed(777)  # for reproducibility
np.random.seed(999)

xy = np.loadtxt("Input_data_path", delimiter=',', dtype=np.float32) #x8, x16

# xy_data=xy.reshape(-1,10) #x8
xy_data=xy.reshape(-1,18) #x16
np.random.shuffle(xy_data)

x_data_training = xy_data[:4500, :-2]
y_data_training = xy_data[:4500, -2]
x_data_test = xy_data[4500:, :-2]
y_data_test = xy_data[4500:, -2:]

forest_reg = RandomForestRegressor(random_state=0, n_estimators=10, n_jobs=-1)
forest_reg.fit(x_data_training, y_data_training)
y_pred_tr=forest_reg.predict(x_data_training)
y_pred=forest_reg.predict(x_data_test)

print('{:.4f}'.format(r2_score(y_data_training, y_pred_tr, multioutput='variance_weighted')))
print('{:.4f}'.format(r2_score(y_data_test, y_pred, multioutput='variance_weighted')))
print('{:.4f}'.format(mean_squared_error(y_data_training, y_pred_tr, multioutput='uniform_average'))) 
print('{:.4f}'.format(mean_squared_error(y_data_test, y_pred, multioutput='uniform_average'))) 

#randomforest_regression
```

```python
import tensorflow as tf
import numpy as np
import random
from sklearn.metrics import r2_score
from sklearn.multioutput import MultiOutputRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error

tf.reset_default_graph()  
tf.set_random_seed(777)  # for reproducibility
np.random.seed(999)

xy = np.loadtxt("Input_data_path", delimiter=',', dtype=np.float32) #x8, x16

# xy_data=xy.reshape(-1,10) #x8
xy_data=xy.reshape(-1,18) #x16
np.random.shuffle(xy_data)

x_data_training = xy_data[:4500, :-2]
y_data_training = xy_data[:4500, -2]
x_data_test = xy_data[4500:, :-2]
y_data_test = xy_data[4500:, -2:]

svr_reg = MultiOutputRegressor(SVR(gamma='scale'),n_jobs=-1)
svr_reg.fit(x_data_training, y_data_training)
y_pred_tr=svr_reg.predict(x_data_training)
y_pred=svr_reg.predict(x_data_test)

print('{:.4f}'.format(r2_score(y_data_training, y_pred_tr, multioutput='variance_weighted')))
print('{:.4f}'.format(r2_score(y_data_test, y_pred, multioutput='variance_weighted')))
print('{:.4f}'.format(mean_squared_error(y_data_training, y_pred_tr, multioutput='uniform_average'))) 
print('{:.4f}'.format(mean_squared_error(y_data_test, y_pred, multioutput='uniform_average'))) 

#SVM_regression 
```

```python
import tensorflow as tf
import numpy as np
import random

from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

tf.reset_default_graph()  
tf.set_random_seed(777)  # for reproducibility
np.random.seed(999)

xy = np.loadtxt("Input_data_path", delimiter=',', dtype=np.float32) #x8, x16

# xy_data=xy.reshape(-1,10) #x8
xy_data=xy.reshape(-1,18) #x16
np.random.shuffle(xy_data)

x_data_training = xy_data[:4500, :-2]
y_data_training = xy_data[:4500, -2]
x_data_test = xy_data[4500:, :-2]
y_data_test = xy_data[4500:, -2:]

gpr = GaussianProcessRegressor(random_state=0,alpha=1e-4, n_restarts_optimizer=0).fit(x_data_training, y_data_training)
gpr_tr, std_tr=gpr.predict(x_data_training, return_std=True)
gpr_, std_=gpr.predict(x_data_test, return_std=True)

print('{:.4f}'.format(mean_squared_error(y_data_training, gpr_tr, multioutput='uniform_average'))) 
print('{:.4f}'.format(mean_squared_error(y_data_test, gpr_, multioutput='uniform_average'))) 
print('{:.4f}'.format(r2_score(y_data_training, gpr_tr, multioutput='variance_weighted'))) 
print('{:.4f}'.format(r2_score(y_data_test, gpr_, multioutput='variance_weighted'))) 

#Gaussian_process_regression
```
